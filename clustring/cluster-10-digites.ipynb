```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
import sklearn.model_selection as select
import sklearn.datasets as dataset
from sklearn.decomposition import PCA
```

```python
X , y_true = dataset.load_digits(n_class=10, return_X_y=True)
xtrain,xtest,ytrain,ytest = select.train_test_split(X , y_true,test_size=0.009,random_state=10,stratify=y_true)
model = KMeans(10)
model.fit(xtrain)
y_pred = model.predict(xtrain)
centers = model.cluster_centers_
```

```python
out_sample = model.predict(xtest)
out = model.cluster_centers_[out_sample]
out_sample
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(xtrain)

# 4) نمایش داده‌ها با رنگ بر اساس خوشه‌ها
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_pred, cmap="tab10", s=15)
plt.title("Digits dataset clustered with KMeans")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar(label="Cluster")
plt.show()

# 5) نمایش مراکز خوشه‌ها (بازسازی به شکل تصویر)
fig, axes = plt.subplots(2, 5, figsize=(5, 2))
centers = centers.reshape(10, 8, 8)
for ax, center, i in zip(axes.ravel(), centers, range(10)):
    ax.imshow(center, cmap="gray",interpolation="nearest")
    ax.set_title(f"Cluster {i}")
    ax.axis("off")
#plt.suptitle("Cluster centers (as images)")
plt.show()
```
