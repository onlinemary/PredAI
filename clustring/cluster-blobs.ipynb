```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
import numpy as np
from sklearn.datasets import make_blobs
```

```python
X , y_true = make_blobs(n_samples=40000, centers=10 ,cluster_std=0.8,random_state=0)
```

```python
plt.scatter(X[:,0],X[:,1],s=10)
plt.show()
```

```python
model = KMeans(n_clusters=10)
model.fit(X)
out_pred = model.predict(X)
centers = model.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1],c=out_pred,s=10,cmap="spring")
plt.scatter(centers[:,0],centers[:,1],s=200,cmap="orange")
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=y_true,s=10,cmap="winter")
plt.scatter(centers[:,0],centers[:,1],s=200,cmap="green")
plt.show()
```
