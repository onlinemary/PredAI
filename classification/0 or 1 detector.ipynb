```python
import matplotlib.pyplot as plt
import sklearn.linear_model as slm
import sklearn.model_selection as select
import sklearn.metrics as metrics
from sklearn import datasets
import numpy as np
import random as rand
```

```python
digites = datasets.load_digits()
rd = rand.randint(0,1790)
print(digites.target[rd])
img = digites.images[rd]
plt.imshow(img)
length = digites.target.shape
```

```python

new_digites_data = []
new_digites_target = []
for i in range(length[0]):
    if (digites.target[i] == 0) or (digites.target[i] == 1):
        print(digites.target[i])
        new_digites_data.append(digites.data[i])
        new_digites_target.append(digites.target[i])
        
```

```python
x = new_digites_data
y = new_digites_target
x_train, x_test , y_train, y_test = select.train_test_split(x , y , shuffle=True, train_size=0.05)
len(x_train)
```

```python
model = slm.LogisticRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
er = y_pred - y_test
er
```

```python
er = y_test - y_pred
er
pred_True = 0
pred_False = 0
for i in range(0,len(er)):
    if (er[i] == 0):
      pred_True = pred_True + 1
    elif (er[i] !=0):
      pred_False = pred_False + 1

print("True =" + str(pred_True) )
print("false = " + str(pred_False))
```
