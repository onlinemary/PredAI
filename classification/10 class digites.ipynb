```python
import matplotlib.pyplot as plt
import sklearn.linear_model as slm
import sklearn.model_selection as select
import sklearn.metrics as metrics
from sklearn import datasets
import numpy as np
```

```python
digites = datasets.load_digits()
x = digites.data
y = digites.target
len(x)
```

```python
x_train, x_test, y_train, y_test = select.train_test_split(x , y , shuffle = True, test_size = 0.2)
len(x_test)
```

```python
model = slm.LogisticRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
er = y_test - y_pred
er
pred_True = 0
pred_False = 0
for i in range(0,len(er)):
    if (er[i] == 0):
      pred_True = pred_True + 1
    elif (er[i] !=0):
      pred_False = pred_False + 1

print("True =" + str(pred_True) )
print("false = " + str(pred_False))
```
